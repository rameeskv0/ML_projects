```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn import metrics
import matplotlib.pyplot as plt
```

```python
x1 = np.array([1,2,3,4,5,5,4,2,2,2,5,2,1,5])
x2 = np.array([1,1,3,4,2,3,2,4,3,5,5,3,2,3])
plt.plot(x1,x2,'ro')
plt.show()
```

```python
x = np.array([x1,x2]).T
```

```python
tests = [2,3,4,5,8]
colors = ['b','g','r','c','m','y','k','b']
markers = ['o','s','D','v','^','v','*','+']
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
subplot_counter = 1
for t in tests:
    plt.tight_layout()
    plt.subplot(3,2,subplot_counter)
    model = KMeans(n_clusters=t).fit(x)
    print(model.labels_)
    for i,l in enumerate(model.labels_):
        plt.plot(x1[i],x2[i],color=colors[l],marker=markers[l])
    plt.xlim([0,10])
    plt.ylim([0,10])
    plt.title('k=%s,coeff=%0.3f'%(t,metrics.silhouette_score(x,model.labels_,metric='euclidean')))
    subplot_counter += 1
```
