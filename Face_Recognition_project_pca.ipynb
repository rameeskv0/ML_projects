```python
from os import walk,path
from sklearn.model_selection import train_test_split
import mahotas as mh
from sklearn.preprocessing import scale
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
x = []
y = []
```

```python
for dir_path,dir_name,file_name in walk("C:/Users/0rame\Desktop/ML_tutorials/Datas/att_faces-20240123T095327Z-001/att_faces"):
    for fn in file_name:
        if fn[-3:] == 'pgm':
            image_file_name = path.join(dir_path,fn)
        image = mh.imread(image_file_name)
        scaled_image = scale(image.reshape(10304).astype('float32'))
        x.append(scaled_image)
        y.append(dir_path)
```

```python
import numpy as np
```

```python
x = np.array(x)
```

```python
x.shape
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y)
```

```python
pca = PCA()
```

```python
reduced_x = pca.fit_transform(xtrain)
```

```python
pc_values = np.arange(pca.n_components_)
```

```python
pc_values
```

```python
pca.explained_variance_ratio_
```

```python
plt.plot(pc_values,pca.explained_variance_ratio_,'o-',linewidth=2,color='b')
plt.title('scree_plot')
plt.xlabel('principle_compenents')
plt.ylabel('variance_explained')
plt.show()
```
