```python
import skimage.io as io
import numpy as np
import matplotlib.pyplot as plt
from skimage.feature import corner_harris,corner_peaks
from skimage.exposure import equalize_hist
from skimage.color import rgb2gray
```

```python
mandrill1 = io.imread("C:/Users/0rame/Desktop/ML_tutorials/Datas/mandrill.png")
```

```python
mandrill2 = equalize_hist(mandrill1)
plt.imshow(mandrill2)
```

```python
mandrill1 = rgb2gray(mandrill2)
plt.gray()
plt.imshow(mandrill1)
```

```python
corners = corner_peaks(corner_harris(mandrill1),min_distance=6)
corners
```

```python
def show_corners(corners,images):
    fig = plt.figure()
    plt.gray()
    plt.imshow(images)
    y_corner,x_corner = zip(*corners)
    plt.plot(x_corner,y_corner,'or')
    plt.xlim(0,images.shape[1])
    plt.ylim(images.shape[0],0)
    fig.set_size_inches(np.array(fig.get_size_inches())*1.5)
    plt.show()
    
```

```python
show_corners(corners,mandrill1)
```
